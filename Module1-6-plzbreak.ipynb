```python
#variable storing types of people as 10 
types_of_people = 10 
#variable for x that stores a sentence saying how many types of people there are 
x = f"There are {types_of_people} types of people."
#variable for binary to equal the string binary 
binary = "binary"
#variable that stores 'don't' string
do_not = "don't"
#variable where y stores an f string saying those who know binary and those who don't 
y = f"Those who know {binary} and those who {do_not}."

#prints the x variable 
print(x)
#prints the y variable 
print(y)

#prints an f-string that includes the value of the variable x
print(f"I said: {x}")
#another f'string except now a y variable
print(f"I also said: '{y}'")

#variable that's a boolean for hilarious to equal false 
hilarious = False
#variable that sets joke evaluation as a string with curly bracket space holder for format() method
joke_evaluation = "Isn't that joke so funny?! {}"

#prints the joke evaluation variable with the 'hilarious' variable added into the format string
print(joke_evaluation.format(hilarious))

#'w' variable set for string to say "this is the left side of'
w = "This is the left side of..."
#'e' variable set for string to say "a string with a right side'
e = "a string with a right side."

#prints the 'w' variable, then the 'e' variable
print(w + e)
```
